# Code - Part 1
### Analyse the data 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

df_health = pd.read_csv('all_data.csv')

# Print the first 5 rows
print(df_health.head())
```

## Data per country

```python
# Getting the countries list
countries = df_health['Country'].unique()
print(countries)
```

```python
# Finding the values by country
df_health_country_grouped = df_health.groupby('Country')

# Create a dictionary of DataFrames
# The keys of the dictionary will be the unique values from the 'Country' column
df_health_country = {country_name: country_data for country_name, country_data in df_health_country_grouped}

# Print the first 5 rows of every country
for country in countries:
    print(df_health_country[country].head())
    print(" ---------- ")
```

## Analyse Country data

Now lets analyse each country individaully 

```python
def generateCountryGraph(df, axes, country):
    sns.scatterplot(x='GDP', y='Life expectancy at birth (years)', data=df, ax=axes)

    axes.set_title("Country =" + country)
    axes.set_xlabel("GDP")
    axes.set_ylabel("Life expectancy")
```

```python
print(" ---------- ")
fig, axes = plt.subplots(2, 3, figsize=(10, 5)) 
graphCount = 0
lineIndex = 0
for country in countries:
    generateCountryGraph(df_health_country[country], axes[lineIndex][graphCount], country)
    graphCount += 1
    if graphCount == 3:
        graphCount = 0
        lineIndex += 1

plt.tight_layout()
plt.show()
plt.clf()
print(" ---------- ")
```

```python
def generateCountryGraphGDP_YEAR(df, axes, country):
    sns.lineplot(x='Year', y='GDP', data=df, ax=axes)

    axes.set_title("Country =" + country)
    axes.set_ylabel("GDP")
    axes.set_xlabel("Year")
```

```python
print(" ---------- ")
fig, axes = plt.subplots(2, 3, figsize=(10, 5)) 
graphCount = 0
lineIndex = 0
for country in countries:
    generateCountryGraphGDP_YEAR(df_health_country[country], axes[lineIndex][graphCount], country)
    graphCount += 1
    if graphCount == 3:
        graphCount = 0
        lineIndex += 1

plt.tight_layout()
plt.show()
plt.clf()
print(" ---------- ")
```

```python
def generateCountryGraphLIFE_YEAR(df, axes, country):
    sns.lineplot(x='Year', y='Life expectancy at birth (years)', data=df, ax=axes)

    axes.set_title("Country =" + country)
    axes.set_ylabel("Life expectancy at birth (years)")
    axes.set_xlabel("Year")
```

```python
print(" ---------- ")
fig, axes = plt.subplots(2, 3, figsize=(10, 5)) 
graphCount = 0
lineIndex = 0
for country in countries:
    generateCountryGraphLIFE_YEAR(df_health_country[country], axes[lineIndex][graphCount], country)
    graphCount += 1
    if graphCount == 3:
        graphCount = 0
        lineIndex += 1

plt.tight_layout()
plt.show()
plt.clf()
print(" ---------- ")
```
