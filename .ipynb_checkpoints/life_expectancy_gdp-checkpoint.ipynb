 GDP and Life Expectancy Analysis

## Project Overview
This project investigates the relationship between the **Gross Domestic Product (GDP)** and **Life Expectancy at Birth** across six different nations. The goal is to determine if a higher GDP consistently correlates with longer life spans and to visualize how these metrics have evolved over time using data from the World Health Organization and the World Bank.

This analysis was completed as part of the **Codecademy Data Science and AI Engineering** career path.
